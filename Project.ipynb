# Usecase-10 Load all three XML datasets in Pandas Data frames and display first 5 records.

```python
import xml.etree.ElementTree as et 
```

```python
xml_pro = et.parse('D:\\Python\\PythonProject\\PythonProject\\XML_s4\\products.xml')
xml_cust = et.parse('D:\\Python\\PythonProject\\PythonProject\\XML_s4\\customers.xml')
xml_tra = et.parse('D:\\Python\\PythonProject\\PythonProject\\XML_s4\\transactions.xml')
```

```python
root = xml_pro.getroot()
root1 = xml_cust.getroot()
root2 = xml_tra.getroot()
```

```python
print(root) 
print(root1)
print(root2)
```

```python
df = pd.DataFrame(columns=["pno","name","isactive","describe","createdDate","UpdatedDate"]) 
df1 = pd.DataFrame(columns = ["custno","firstname","lastname","gender","age","profession","contactNo","emailId","city","state","isActive","createdDate","UpdatedDate"])
df2 = pd.DataFrame(columns= ["txnno","txndate","custno","amount","productno","spendby"])
```

```python
for tag in root:
    
    productno = tag.find("productno").text
    productName = tag.find("productName").text
    isActive = tag.find("isActive").text
    Description = tag.find("Description").text
    createdDate = tag.find("createdDate").text
    UpdatedDate = tag.find("UpdatedDate").text
    df = df.append(pd.Series([productno,productName,isActive,Description,createdDate,UpdatedDate],index=["pno","name","isactive","describe","createdDate","UpdatedDate"]),ignore_index=True)
```

```python
for tags in root1:
    
    custno = tags.find("custno").text 
    fstname = tags.find("firstname").text
    lstname = tags.find("lastname").text
    gender = tags.find("gender").text
    age = tags.find("age").text
    prof = tags.find("profession").text
    contno = tags.find("contactNo")
    email = tags.find("emailId").text
    city = tags.find("city").text
    state = tags.find("state").text
    isActive = tags.find("isActive").text
    createdDate = tags.find("createdDate").text
    Updateddate = tags.find("UpdatedDate").text
    df1 = df1.append(pd.Series([custno,fstname,lstname,gender,age,prof,contno,email,city,state,isActive,createdDate,Updateddate],index =["custno","firstname","lastname","gender","age","profession","contactNo","emailId","city","state","isActive","createdDate","UpdatedDate"]),ignore_index=True)
    
    
    
```

```python
for tag2  in root2:
    txnn = tag2.find("txnno").text
    txndat = tag2.find("txndate").text
    custn = tag2.find("custno").text
    amoun = tag2.find("amount").text
    productn = tag2.find("productno").text
    spendb = tag2.find("spendby").text
    df2 = df2.append(pd.Series([txnn,txndat,custn,amoun,productn,spendb],index = ["txnno","txndate","custno","amount","productno","spendby"]),ignore_index =True)
    
```
